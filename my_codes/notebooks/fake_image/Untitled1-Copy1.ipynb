```python
# name = ' '.join(n)
os.makedirs("img2")
for n in range(27):
    i = Image.open("bg1.png")
    draw = ImageDraw.Draw(i)
    font = ImageFont.truetype("newfont/07YasashisaAntique.ttf", 20)
    text = f"{randint(2024, 2040)}年(令和{randint(1, 12):02d}年){randint(1, 12):02d}月{randint(1, 32):02d}日まで有効"
    draw.text((4, 5),text,(0,0,0),font=font)
    i.save(f"img2/{n}.png")
    with open("labels_2.txt", "a") as f:
        f.write(f"{text}\n")
#     plt.imshow(i)
```
