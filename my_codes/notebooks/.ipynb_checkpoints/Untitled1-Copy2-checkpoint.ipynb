```python
%cd /home/gmo/Downloads/fake_data
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from PIL import ImageDraw, ImageFont, ImageEnhance, ImageFilter
import os
from random import randint
import random
from gimei import Gimei
import pickle as pkl
```

```python
mpl.rcParams['figure.figsize'] = (20, 20)
```

```python
def path2np_img(path):
    i = Image.open(path)
    return np.array(i)
```

```python
def gen_name():
    name = Gimei().name
    lst = [name.first.kanji, name.first.hiragana, name.first.katakana]
    return random.choice(lst)
```

```python
mode = 'save'
path = 'res/img3_2'
os.makedirs(path, exist_ok=True)
# for n in range(len(names)):
l_i_2 = []
for n in range(1, 49*50*4 + 1):
    i_empty = np.full((53 * 3, 565 * 3, 3), 255)
    i_empty[:, :10] = 80
    i_empty[:, -10:] = 80
    i_empty[:10, :] = 80
    i_empty[:, :5] = 40
    i_empty[:, -5:] = 40
    i_empty[:5, :] = 40
    i_empty[-5:, :] = 40

    i = Image.fromarray(i_empty.astype("uint8"))

    draw = ImageDraw.Draw(i)
    font = ImageFont.truetype("newfont/meiryo.ttc", 110)
#     name = names[n]
    name = gen_name()[:4]
    if len(name)==1:
        name_pad = " ".join(name)
        text = f"     日 本{name_pad:^6}[東 京 {name_pad:^4}]"
    elif len(name)==2:
        name_pad = " ".join(name)
        text = f"日 本{name_pad:^10}[東 京 {name_pad:^7}]"
    elif len(name)==3:
        name_pad = " ".join(name)
        text = f"日 本{name_pad:^9}[東 京 {name_pad}]"
    elif len(name)==4:
        name_pad = name
        text = f"日 本{name_pad:^6}[東 京 {name_pad:^}]"
    draw.text((70, -5),text,(30,30,30),font=font)
    
    l_i_2.append(i)
    
    if n % (49*4) == 0 :
        os.makedirs(f"{path}_a4", exist_ok=True)
        images_iter = iter(l_i_2)
        h, w = 49, 520
        pad = np.full((20, w, 3), 255, 'uint8')
        col1, col2, col3, col4 = [], [], [], []
        cols = [col1, col2, col3, col4]
        for ik in cols:
            for j in range(49):
                img = next(images_iter)
                k = img.resize((w, h), Image.ANTIALIAS)
                k = np.array(k)
                ik.extend((k, pad))
        colskk = list(map(np.vstack, cols))

        bg = np.full((3508,2480,3), 255, dtype='uint8')
        coords = [[90+590*i, 80] for i in range(len(colskk))]
        for colkk, coord in zip(colskk, coords):
            x, y = coord
            bg[y:y+colkk.shape[0], x:x+colkk.shape[1]] = colkk
        Image.fromarray(bg).save(f"{path}_a4/{int(n / (49*4)):03d}.png")
        l_i_2 = []
    
    if mode=='plot':
        plt.imshow(i)
        break
    elif mode=='save':
        i.save(f"{path}/{n:03d}_{name}.png")
        with open(f"{path}/{n:03d}_{name}.txt", "w") as f:
            f.write(f"{name}")
```

```python
with open(f"{path}/all_images.pkl", "wb") as f:
    pkl.dump(l_i_2, f)
len(l_i_2)

# with open(f"{path}/all_images.pkl", "rb") as f:
#     l_i_2 = pkl.load(f)
# len(l_i_2)
```

```python
mode = 'save'
path = "res/img4_2"
os.makedirs(path, exist_ok=True)
l_i = []
for n in range(1, 44*3*50 + 1):
    year = randint(2000, 2040)
    h_year = f"令和{year-2019+1:02d}" if year > 2019 else f"平成{year-1990+1:02d}"
    text = f"{year}年({h_year}年){randint(1, 12):02d}月{randint(1, 32):02d}日まで有効"

    i_empty = np.full((190, 2800, 3), 255)
    i = Image.fromarray(i_empty.astype('uint8'))
    draw = ImageDraw.Draw(i)
    font = ImageFont.truetype("fonts/Font/儷黑 Pro.ttf", 155)
    draw.text((50, 15),text,(0,0,0),font=font)
    
    i = i.resize((1879, 180), Image.ANTIALIAS)
    
    enhancer = ImageEnhance.Sharpness(i)
    factor = 1 / 4.0
    i = enhancer.enhance(factor)
    
    body = np.array(i)

    bg = path2np_img("bg/bg5.png")
    bg[np.where(body<130)] = body[np.where(body<130)]
    
    i = Image.fromarray(bg.astype('uint8'))
    i = i.filter(ImageFilter.SMOOTH)
    
    l_i.append(i)
    
    if n % (44*3) == 0 :
        os.makedirs(f"{path}_a4", exist_ok=True)
        images_iter = iter(l_i)
        h, w = 57, 602
        pad = np.full((20, w, 3), 255, 'uint8')
        col1, col2, col3 = [], [], []
        cols = [col1, col2, col3]
        for ik in cols:
            for ij in range(44):
                img = next(images_iter)
                k = img.resize((w, h), Image.ANTIALIAS)
                k = np.array(k)
                ik.extend((k, pad))
        colskk = list(map(np.vstack, cols))

        bg = np.full((3508,2480,3), 255, dtype='uint8')
        coords = [[200+750*i, 80] for i in range(len(colskk))]
        for colkk, coord in zip(colskk, coords):
            x, y = coord
            bg[y:y+colkk.shape[0], x:x+colkk.shape[1]] = colkk
        Image.fromarray(bg).save(f"{path}_a4/{int(n / (44*3)):03d}.png")
        l_i = []

    if mode=='plot':
        plt.imshow(i)
        break
    elif mode=='save':
#         pass
        i.save(f"{path}/{n:03d}_{text}.png")
        with open(f"{path}/{n:03d}_{text}.txt", "w") as f:
            f.write(f"{text}")
```
